```python
#<aside>
üí° **Question 1**

Given an integer `n`, return *`true` if it is a power of two. Otherwise, return `false`*.

An integer `n` is a power of two, if there exists an integer `x` such that `n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

</aside>

"""To determine if an integer is a power of two, we can use the following approach:

   1. Check if the given integer n is less than or equal to 0. If it is, return false because 0 and negative numbers cannot
      be powers of two.

   2. Check if n is equal to 1. If it is, return true because 1 is considered a power of two.

   3. Use a loop to continuously divide n by 2 until it becomes 1 or an odd number. If at any point n becomes an odd number, 
      it cannot be a power of two, so return false.

   4. If the loop completes and n becomes 1, return true because it is a power of two."""

#Here's the implementation of this algorithm in Python:

def is_power_of_two(n):
    if n <= 0:
        return False
    if n == 1:
        return True
    
    while n % 2 == 0:
        n = n // 2
    
    return n == 1
print(is_power_of_two(1))  # Output: True
print(is_power_of_two(16))  # Output: True
print(is_power_of_two(3))  # Output: False

#The function returns the expected outputs for the given examples.
```

```python
#<aside>
üí° **Question 2**

Given a number n, find the sum of the first natural numbers.

**Example 1:**

Input: n = 3 

Output: 6

**Example 2:**

Input  : 5 

Output : 15

</aside>

"""To find the sum of the first n natural numbers, we can use the formula for the sum of an arithmetic series:

   Sum = (n / 2) * (first_number + last_number)
   
   In this case, the first number is 1 and the last number is n. So, we can substitute these values into the formula:
   
   Sum = (n / 2) * (1 + n)"""

#Here's the implementation of this algorithm in Python:

def sum_of_natural_numbers(n):
    return (n / 2) * (1 + n)
print(sum_of_natural_numbers(3))  # Output: 6
print(sum_of_natural_numbers(5))  # Output: 15

#The function returns the expected outputs for the given examples.
```

```python
#<aside>
üí° **Question 3**

****Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

</aside>

"""To find the factorial of a positive integer N, we can use a simple iterative approach. The factorial of N is the product 
   of all positive integers from 1 to N.

   Here's the algorithm to find the factorial of N:

   2. Initialize a variable factorial to 1.

   3. Use a loop to multiply factorial by numbers from 1 to N.

   4. After the loop completes, the value of factorial will be the factorial of N."""

#Here's the implementation of this algorithm in Python:

def factorial(N):
    factorial = 1
    for i in range(1, N + 1):
        factorial *= i
    return factorial
print(factorial(5))  # Output: 120
print(factorial(4))  # Output: 24

#The function returns the expected outputs for the given examples.
```

```python
#<aside>
üí° **Question 5**

Given an array of integers¬†**arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

</aside>

"""To find the maximum element in an array arr using recursion, we can divide the problem into subproblems. Here's the 
   algorithm:

   1. If the array contains only one element, return that element as the maximum.

   2. Otherwise, divide the array into two halves: the first half and the second half.

   3. Recursively find the maximum element in each half of the array.

   4. Compare the maximum elements found in the two halves and return the maximum of the two."""

#Here's the implementation of this algorithm in Python:

def find_max(arr, start, end):
    if start == end:
        return arr[start]
    
    mid = (start + end) // 2
    max_first_half = find_max(arr, start, mid)
    max_second_half = find_max(arr, mid + 1, end)
    
    return max(max_first_half, max_second_half)

#We need to pass the array arr, the start index (0), and the end index (length of the array minus 1) to the recursive function.

arr1 = [1, 4, 3, -5, -4, 8, 6]
print(find_max(arr1, 0, len(arr1) - 1))  # Output: 8

arr2 = [1, 4, 45, 6, 10, -8]
print(find_max(arr2, 0, len(arr2) - 1))  # Output: 45

#The function returns the expected outputs for the given examples.
```

```python
#<aside>
üí° **Question 6**

Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth
term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

</aside>

"""To find the nth term of an arithmetic progression (AP) series, we can use the formula:

   nth_term = a + (n - 1) * d
   
   Where:
   . a is the first term of the AP
   . d is the common difference between consecutive terms
   . n is the position of the term we want to find"""

#Here's the implementation of this algorithm in Python:

def find_nth_term(a, d, n):
    nth_term = a + (n - 1) * d
    return nth_term
print(find_nth_term(2, 1, 5))  # Output: 6
print(find_nth_term(5, 2, 10))  # Output: 23

#The function returns the expected outputs for the given examples.
```

```python
#<aside>
üí° **Question 7**

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = ‚ÄúABC‚Äù*

***Output:***

*‚ÄúABC‚Äù, ‚ÄúACB‚Äù, ‚ÄúBAC‚Äù, ‚ÄúBCA‚Äù, ‚ÄúCBA‚Äù, ‚ÄúCAB‚Äù*

**Example 2:**

***Input:***

*S = ‚ÄúXY‚Äù*

***Output:***

*‚ÄúXY‚Äù, ‚ÄúYX‚Äù*

</aside>

"""To generate all permutations of a given string S, we can use a recursive approach. Here's the algorithm:

   1. If the length of the string S is 1, there is only one permutation, which is the string itself. Return a list containing S.

   2. Otherwise, iterate over each character c in S:
      . Remove c from S to create a new string new_S.
      . Recursively generate all permutations of new_S.
      . For each permutation perm of new_S, append c to the beginning of perm to create a new permutation.
      . Add the new permutation to the list of permutations.
      
   4. Return the list of permutations."""

#Here's the implementation of this algorithm in Python:

def permute(S):
    if len(S) == 1:
        return [S]

    permutations = []
    for i in range(len(S)):
        c = S[i]
        new_S = S[:i] + S[i+1:]
        sub_permutations = permute(new_S)
        for perm in sub_permutations:
            permutations.append(c + perm)
    
    return permutations
S1 = "ABC"
print(permute(S1))  # Output: ['ABC', 'ACB', 'BAC', 'BCA', 'CAB', 'CBA']

S2 = "XY"
print(permute(S2))  # Output: ['XY', 'YX']

#The function returns the expected outputs for the given examples.
```

```python
#<aside>
üí° **Question 8**

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

</aside>

"""To find the product of all elements in an array, we can iterate over the array and multiply each element with the running 
   product. Here's the algorithm:

   1. Initialize a variable product to 1.

   2. Iterate over each element num in the array.

   3. Multiply num with the current value of product and update product with the result.

   4. After the loop completes, product will hold the product of all elements in the array."""

#Here's the implementation of this algorithm in Python:

def product_of_array(arr):
    product = 1
    for num in arr:
        product *= num
    return product
arr1 = [1, 2, 3, 4, 5]
print(product_of_array(arr1))  # Output: 120

arr2 = [1, 6, 3]
print(product_of_array(arr2))  # Output: 18

#The function returns the expected outputs for the given examples.
```
